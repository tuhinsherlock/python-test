```python
from nsepy import get_history
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date
```

```python
tcs = get_history(symbol='TCS',
                   start=date(2015,1,1),
                   end=date(2016,12,31))
```

```python
tcs
```

```python
infy=get_history(symbol='INFY',start=date(2015,1,1),end=date(2016,12,31))
```

```python
nifty = get_history(symbol="NIFTYIT",start=date(2015,1,1),end=date(2016,12,31),index=True)
```

```python
nifty
```

```python
def time_series(t,weeks):
    dd=pd.DataFrame()
    dd['Closing']=t['Close'].resample('W').mean()
    for i in weeks:
        moving=dd['Closing'].rolling(i).mean()
        dd['Moving average for ' + str(i) +' weeks']=moving
        print('Moving average for '+str(i)+' weeks of '+t.name+':')
        print(dd['Moving average for ' + str(i) +' weeks'])
    dd.plot()
    plt.title('Moving averages for '+t.name)
    plt.show()
```

```python
weeks=[4,16,28,40,52]
tcs.name='TCS'
infy.name='INFY'
nifty.name='NIFTY IT'
tcs.index=pd.to_datetime(tcs.index)
infy.index=pd.to_datetime(infy.index)
nifty.index=pd.to_datetime(nifty.index)
```

```python
time_series(tcs,weeks)
```

```python
time_series(infy,weeks)
```

```python
time_series(nifty,weeks)
```

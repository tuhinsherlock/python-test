```python
#function to calculate time series rolling window moving average

def time_series_mov_avg(t,weeks): 
    dd=pd.DataFrame()
    dd['Closing']=t['Close'].resample('W').mean()
    for i in weeks:
        moving=dd['Closing'].rolling(i).mean()
        dd['Moving average for ' + str(i) +' weeks']=moving
        print('Moving average for '+str(i)+' weeks of '+t.name+':')
        print(dd['Moving average for ' + str(i) +' weeks'])
    dd.plot()
    plt.title('Moving averages for '+t.name)
    plt.show()
    
```

```python
#names assigned to the stocks and indexes converted to datetime

weeks=[4,16,28,40,52]
tcs.name='TCS'
infy.name='INFY'
nifty.name='NIFTY IT'
tcs.index=pd.to_datetime(tcs.index)
infy.index=pd.to_datetime(infy.index)
nifty.index=pd.to_datetime(nifty.index)
```

```python
time_series_mov_avg(tcs,weeks) #calculating moving average of TCS
```

```python
time_series_mov_avg(infy,weeks) #calculating Moving average of infy
```

```python
time_series_mov_avg(nifty,weeks) #calculating moving average of Nifty
```

```python
#filling the non empty days by forward fill as stock data is carried over on holidays
tcs1=tcs.asfreq('D',method='ffill')
infy1=infy.asfreq('D',method='ffill')
nifty1=nifty.asfreq('D',method='ffill')
day_win=[10,75]
tcs1.name='TCS'
infy1.name='INFY'
nifty1.name='NIFTY IT'
```

```python
#function to find moving average of day based rolling window
def daily_mov_avg(t,day_win):
    dd=pd.DataFrame()
    dd['Closing']=t['Close']
    for i in day_win:
        moving=dd['Closing'].rolling(i).mean()
        dd['Moving Average for '+str(i)+' days']=moving
        print('Moving Average for '+str(i)+' days')
        print(dd['Moving Average for '+str(i)+' days'])
    dd.plot()
    plt.title('Moving Averages for '+t.name)
    plt.show()
```

```python
daily_mov_avg(tcs1,day_win) #day-wise moving average for TCS
```

```python
daily_mov_avg(infy1,day_win)  #day-wise moving average for INFY
```

```python
daily_mov_avg(nifty1,day_win)  #day-wise moving average for NIFTY
```

```python
#Function to calculate Volume Shock
def vol_shock(index):
    index['Vol_Shock']=(abs(index['Volume'].pct_change().mul(100))>10).astype(int)
    return index
```

```python
tcs11=tcs1
nifty11=nifty1
infy11=infy1
```

```python
#Calculating Volume Shock for TCS, INFY, NIFTY
tcs1=vol_shock(tcs1)
infy1=vol_shock(infy1)
nifty1=vol_shock(nifty1)
```

```python
#Functions to calculate Volume Shock Direction
def vol_dir(index):
    if index['Vol_Shock']==0:
        return 0
    else:
        if (index['Vol_Shift']-index['Volume'])>0:
            return 1
        else:
            return -1
def vol_shock_dir(index):
    index['Vol_Shift']=index['Volume'].shift(1)
    index['Vol_Shock_Dir']=index.apply(vol_dir,axis=1)
    return index
```

```python
#Calculating Volume Shock Direction
tcs1=vol_shock_dir(tcs1)
infy1=vol_shock_dir(infy1)
nifty1=vol_shock_dir(nifty1)
```

```python
#Function to calculate Price Shock and Price Black Swan, as they are the same thing
def price_shock(index):
    index['Price_Shock']=(abs(index['Close'].pct_change().mul(100))>2).astype(int)
    index['Price_Black_Swan']=index['Price_Shock']
    return index
```

```python
#Calculating Price Shock and Price Black Swan for TCS,INFY,NIFTY
tcs1=price_shock(tcs1)
infy1=price_shock(infy1)
nifty1=price_shock(nifty1)
```

```python
#Functions to calculate Price Shock Direction
def price_dir(index):
    if index['Price_Shock']==0:
        return 0
    else:
        if (index['Price_Shift']-index['Close'])>0:
            return 1
        else:
            return -1
def price_shock_dir(index):
    index['Price_Shift']=index['Close'].shift(1)
    index['Price_Shock_Dir']=index.apply(price_dir,axis=1)
    return index
```

```python
#Calculating Price Shock Direction
tcs1=price_shock_dir(tcs1)
infy1=price_shock_dir(infy1)
nifty1=price_shock_dir(nifty1)
```

```python
#Functions to calculate Price Shock without Volume Shock
def price_vol_check(index):
    if index['Price_Shock']==1 and index['Vol_Shock']==0:
        return 1
    else:
        return 0
def price_without_vol(index):
    index['Price_Without_Vol']=index.apply(price_vol_check,axis=1)
    return index
```

```python
#Calculating Price Shock without Volume Shock for TCS, INFY, NIFTY
tcs1=price_without_vol(tcs1)
infy1=price_without_vol(infy1)
nifty1=price_without_vol(nifty1)
```

```python
#Calculating 52 week moving average for TCS and INFY
test=pd.DataFrame()
dd=pd.DataFrame()
dd['Closing']=tcs1['Close'].resample('W').mean()
test['52_TCS']=dd['Closing'].rolling(52).mean()
test.index=dd.index
dd['Closing']=infy1['Close'].resample('W').mean()
test['52_INFY']=dd['Closing'].rolling(52).mean()
test['Diff']=test['52_TCS'].sub(test['52_INFY'])
test=test.dropna()
```

```python
#plotting time series plot and marking the price without volume shock in the line plot
def plotting(index):
    p=figure(plot_width=600,plot_height=500,title='time series for {}'.format(index.name),x_axis_type='datetime')
    p.line(index.index,index['Close'],color='blue')
    p.circle(index.index, index['Close']*index['Price_Without_Vol'], legend='Price Shock Without Vol Shock',color='red')
    show(p)
```

```python
plotting(tcs1)
plotting(infy1)
output_file('Plotting.html')
```

```python
#plotting difference of 52 week moving average with color in blue spectrum
color_mapper = LinearColorMapper(palette='Blues9', low=min(test['Diff']), high=max(test['Diff']))
p = figure(plot_width=800,plot_height=500,title='Difference in 52 week moving average',x_axis_type='datetime')
p.scatter(test.index,test['Diff'],color={'field': 'y', 'transform': color_mapper})
show(p)
```

```python
from statsmodels.tsa.stattools import pacf

def generate_pacf(t):
    
    a = 50

    x = list(range(a))

    p = figure(plot_height=800, title="Partial Autocorrelation Plot for {}" .format(t.name))

    partial_autocorr = pacf(t["Close"], nlags=a)
    p.vbar(x=x, top=partial_autocorr, width=0.9)
    show(p)
```

```python
output_file("PACF_Plot.html")
generate_pacf(tcs1)
generate_pacf(infy1)
generate_pacf(nifty1)
```
